# Experiment 7: PDE Accuracy

We aim to solve a 2D poisson equation $\nabla^2 f(x,y) = -2\pi^2{\rm sin}(\pi x){\rm sin}(\pi y)$, with boundary condition $f(-1,y)=f(1,y)=f(x,-1)=f(x,1)=0$. The ground truth solution is $f(x,y)={\rm sin}(\pi x){\rm sin}(\pi y)$.

**Objective**: Further experiments into PDE solving on KAN.

## Step 1: PDE

The goal is to solve the second-order elliptic partial differential equation (Laplace's equation) over a square domain $\Omega$:

$$
-\Delta u(x, y) = f(x, y), \quad (x, y) \in \Omega, \quad \Omega = [-1, 1] \times [-1, 1],
$$

where $$\Delta$$ is the Laplace operator:

$$
\Delta u = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2}.
$$

The exact solution is:

$$
u(x, y) = \sin(\pi x) \sin(\pi y),
$$

and the corresponding source term is:

$$
f(x, y) = -2\pi^2 \sin(\pi x) \sin(\pi y).
$$

---

## Step 2: Initialization Process

### 1. Domain Discretization

- **Internal Grid Points**: $x_i$ are generated using a meshgrid over the domain $\Omega$$with resolution $n_i \times n_i$.
- **Boundary Grid Points**: $x_b$ are extracted from the domain boundaries $\partial \Omega$:
  - Top, bottom, left, and right edges.

### 2. KAN Model Initialization

- **Input Dimension**: $2$ (coordinates $x, y$).
- **Output Dimension**: $1$ (scalar solution $u(x, y)$).
- **Hidden Layer Grid**: Adaptive grid refinement with grid sizes $5 \to 10 \to 20$.
- **Basis Function Order**: $k = 3$ (cubic splines).

### 3. Loss Function

The total loss function $\mathcal{L}$ combines the PDE residual loss $\mathcal{L}_{\text{PDE}}$ and the boundary condition loss $\mathcal{L}_{\text{BC}}$:

$$
\mathcal{L} = \alpha \cdot \mathcal{L}_{\text{PDE}} + \mathcal{L}_{\text{BC}},
$$

where:

- **PDE Residual Loss**:

$$
\mathcal{L}_{\text{PDE}} = \frac{1}{N_i} \sum_{i=1}^{N_i} \left( \Delta u_{\theta}(x_i) - f(x_i) \right)^2,
$$

- **Boundary Condition Loss**:

$$
\mathcal{L}_{\text{BC}} = \frac{1}{N_b} \sum_{i=1}^{N_b} \left( u_{\theta}(x_b) - u_{\text{true}}(x_b) \right)^2.
$$

---

## Step 3: Training Workflow

The training process involves the following steps:

1. **Initialize the KAN model** with a coarse grid.
2. **Optimize the model parameters** using the LBFGS optimizer for a fixed number of steps.
3. During training:
   - Compute the PDE residual loss $\mathcal{L}_{\text{PDE}}$at internal points $x_i$.
   - Compute the boundary loss $\mathcal{L}_{\text{BC}}$ at boundary points $x_b$.
   - Update the grid adaptively every 5 steps in the early training stage.
4. **Refine the grid resolution** (e.g., $5 \to 10 \to 20$) for improved accuracy.
5. **Repeat optimization** until the model converges.

---


```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from tensorkan.MultKAN import MultKAN as KANTF
from tensorkan.LBFGS import LBFGS


device = '/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'
print(f"Using device: {device}")

dim = 2
np_i = 51  
np_b = 51  
ranges = [-1, 1]

# Define batch Jacobian using tf.GradientTape

def batch_jacobian(func, x, create_graph=False):
    """
    Computes the Jacobian matrix for a batch of inputs.

    Args:
        func: Callable
            A function mapping x to its output y.
        x: tf.Tensor
            Input tensor of shape (Batch, Length).
        create_graph: bool
            Whether to create a graph for further differentiation.

    Returns:
        jacobian: tf.Tensor
            The Jacobian matrix of shape (Batch, Length, Output_Dim).
    """
    def _func_sum(x):
        """
        Helper function to sum the outputs of `func`.
        This ensures compatibility with `tf.GradientTape`.
        """
        return tf.reduce_sum(func(x), axis=0)

    # Compute the Jacobian
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(x)
        summed_outputs = _func_sum(x)  

    
    raw_jacobian = tape.jacobian(summed_outputs, x, experimental_use_pfor=False)

    
    jacobian = tf.transpose(raw_jacobian, perm=[1, 0, 2])

    return jacobian

# Define solution
sol_fun = lambda x: tf.sin(np.pi * x[:, 0:1]) * tf.sin(np.pi * x[:, 1:2])
source_fun = lambda x: -2 * np.pi**2 * tf.sin(np.pi * x[:, 0:1]) * tf.sin(np.pi * x[:, 1:2])


x_mesh = np.linspace(ranges[0], ranges[1], num=np_i, dtype=np.float32)
y_mesh = np.linspace(ranges[0], ranges[1], num=np_i, dtype=np.float32)
X, Y = np.meshgrid(x_mesh, y_mesh, indexing="ij")

x_i = tf.transpose(
    tf.stack([tf.reshape(X, [-1]), tf.reshape(Y, [-1])]), perm=[1, 0]
)


helper = lambda X, Y: tf.transpose(
    tf.stack([tf.reshape(X, [-1]), tf.reshape(Y, [-1])]), perm=[1, 0]
)


xb1 = helper(X[0, :], Y[0, :])      
xb2 = helper(X[-1, :], Y[-1, :])     
xb3 = helper(X[:, 0], Y[:, 0])       
xb4 = helper(X[:, -1], Y[:, -1])     


x_b = tf.concat([xb1, xb2, xb3, xb4], axis=0)  




alpha = 0.01
log = 1
grids = [5, 10, 20]
steps = 50


pde_losses = []
bc_losses = []
l2_losses = []





for grid in grids:
    if grid == grids[0]:
        
        model = KANTF(width=[2, 2, 1], grid=grid, k=3, seed=1)
        model = model.speed()
    else:
        
        model.save_act = True
        model.get_act(x_i)
        model = model.refine(grid)
        model = model.speed()

    def train():
       
        optimizer = LBFGS(
            learning_rate=1.0,
            max_iter=steps,
            history_size=10,
            line_search_fn="strong_wolfe",
            tolerance_grad=1e-32,
            tolerance_change=1e-32,
        )

        pbar = tqdm(range(steps), desc='description', ncols=100)

        def loss_fn():
            
            sol = sol_fun(x_i)  
            sol_D1_fun = lambda x: batch_jacobian(model, x)[:, 0, :]
            sol_D1 = sol_D1_fun(x_i)
            sol_D2 = batch_jacobian(sol_D1_fun, x_i)  
            lap = tf.reduce_sum(tf.linalg.diag_part(sol_D2), axis=1, keepdims=True)
            source = source_fun(x_i)  
            pde_loss = tf.reduce_mean((lap - source) ** 2)

            
            bc_true = sol_fun(x_b)  
            bc_pred = model(x_b)    
            bc_loss = tf.reduce_mean((bc_pred - bc_true) ** 2)

           
            loss = alpha * pde_loss + bc_loss
            return loss

        for step in pbar:
            
            loss = optimizer.minimize(loss_fn, model.trainable_variables)
            loss = tf.convert_to_tensor(loss, dtype=tf.float32)

            
            sol = sol_fun(x_i)
            sol_D1_fun = lambda x: batch_jacobian(model, x)[:, 0, :]
            sol_D1 = sol_D1_fun(x_i)
            sol_D2 = batch_jacobian(sol_D1_fun, x_i)
            lap = tf.reduce_sum(tf.linalg.diag_part(sol_D2), axis=1, keepdims=True)
            source = source_fun(x_i)
            pde_loss = tf.reduce_mean((lap - source) ** 2)

            bc_true = sol_fun(x_b)
            bc_pred = model(x_b)
            bc_loss = tf.reduce_mean((bc_pred - bc_true) ** 2)
            l2 = tf.reduce_mean((model(x_i) - sol) ** 2)

            
            if step % 5 == 0 and step < 20:
                model.update_grid_from_samples(x_i)

            
            if step % log == 0:
                pbar.set_description(
                    f"loss: {loss.numpy():.2e} | l2: {l2.numpy():.2e}"
                )

            
            pde_losses.append(float(pde_loss))
            l2_losses.append(float(l2))
            bc_losses.append(float(bc_loss))

    
    train()
```

```python
plt.plot(pde_losses, marker='o')
plt.plot(bc_losses, marker='o')
plt.plot(l2_losses, marker='o')
plt.yscale('log')
plt.xlabel('steps')
plt.legend(['PDE loss', 'BC loss', 'L2 squared'])
```
