# Experiment 8: Continual Learning

Our goal is to learn a 1D function from samples. The 1D function has 5 Gaussian peaks. Instead of presenting all samples to NN all at once, we have five phases of learning. In each phase only samples around one peak is presented to KAN. We find that KANs can do continual learning thanks to locality of splines.

**Objective**: Verifying if KAN can do Continual Learning

## Step 1 Construct 5 Gaussian Peaks Dataset

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

datasets = []


n_peak = 5
n_num_per_peak = 100
n_sample = n_peak * n_num_per_peak

x_grid = tf.linspace(-1.0, 1.0, num=n_sample)
x_centers = 2 / n_peak * (np.arange(n_peak) - n_peak / 2 + 0.5) 


x_sample_list = [
    tf.linspace(-1.0 / n_peak, 1.0 / n_peak, num=n_num_per_peak) + center
    for center in x_centers
]
x_sample = tf.reshape(tf.stack(x_sample_list), [-1])


y = 0.0
for center in x_centers:
    y += tf.exp(-300 * (x_grid - center) ** 2)


y_sample = 0.0
for center in x_centers:
    y_sample += tf.exp(-300 * (x_sample - center) ** 2)


plt.plot(x_grid.numpy(), y.numpy(), label="Continuous y")  # 连续曲线
plt.scatter(x_sample.numpy(), y_sample.numpy(), color="red", label="Sampled points")  # 采样点
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gaussian Peaks")
plt.show()
```

## Step 2 Create a figure with 1 row and 5 columns of subplots

```python
# Create a figure with 1 row and 5 columns of subplots, set the figure size
plt.subplots(1, 5, figsize=(15, 2))  
plt.subplots_adjust(wspace=0, hspace=0)  # Remove spacing between subplots

# Loop to create 5 subplots
for i in range(1, 6):  
    plt.subplot(1, 5, i)  
    group_id = i - 1 

    # Plot the continuous curve (Gaussian peaks) with low opacity
    plt.plot(x_grid.numpy(), y.numpy(), color='black', alpha=0.1)

    # Plot the sample points for the current group (subset of x_sample and y_sample)
    plt.scatter(
        x_sample[group_id * n_num_per_peak:(group_id + 1) * n_num_per_peak].numpy(),
        y_sample[group_id * n_num_per_peak:(group_id + 1) * n_num_per_peak].numpy(),
        color="black", s=2  
    )

    # Set x and y axis limits for each subplot
    plt.xlim(-1, 1)  
    plt.ylim(-1, 2)  

# Show the resulting figure
plt.show()
```

## Step 3 Training KAN in a loop, each loop is one subplot

```python
from tensorkan.MultKAN import MultKAN as KANTF

ys = []


model = KANTF(
    width=[1, 1], grid=200, k=3, noise_scale=0.1, 
    sp_trainable=False, sb_trainable=False, base_fun='zero'
)


for group_id in range(n_peak):
    
    dataset = {}
    dataset['train_input'] = tf.reshape(x_sample[group_id * n_num_per_peak:(group_id + 1) * n_num_per_peak], [-1, 1])
    dataset['train_label'] = tf.reshape(y_sample[group_id * n_num_per_peak:(group_id + 1) * n_num_per_peak], [-1, 1])
    dataset['test_input'] = tf.reshape(x_sample[group_id * n_num_per_peak:(group_id + 1) * n_num_per_peak], [-1, 1])
    dataset['test_label'] = tf.reshape(y_sample[group_id * n_num_per_peak:(group_id + 1) * n_num_per_peak], [-1, 1])

   
    model.fit(dataset, opt='LBFGS', steps=50, update_grid=False,lr=1e-5)

    y_pred = model(tf.reshape(x_grid, [-1, 1]))
    ys.append(y_pred.numpy()[:, 0])  
```

## Prediction of KAN in each stage Graph

```python
plt.subplots(1, 5, figsize=(15, 2))
plt.subplots_adjust(wspace=0, hspace=0)

for i in range(1,6):
    plt.subplot(1,5,i)
    group_id = i - 1
    plt.plot(x_grid.numpy(), y.numpy(), color='black', alpha=0.1)
    plt.plot(x_grid.numpy(), ys[i-1], color='black')
    plt.xlim(-1,1)
    plt.ylim(-1,2)
```

## From the graph, we know that kan can do the continue learning
