# Experiment 10: Comparison Between Different Architectures

In this experiment, we'll use several formulas to see their performance, so that we may further evaluate all of them

**Objective**: To evaluate the effect of multiple architectures.

## Step 1: 

$$
f(x) = e^{\sin(\pi x_1) + x_2^2}
$$

```python
from tensorkan.utils import create_dataset
import numpy as np
from tensorkan.MultKAN import MultKAN as KANTF
import tensorflow as tf
#from tensorkan.KANLayer import KANLayer  # 假设 KANLayer 在这个路径下

model = KANTF(width=[2, 5, 1], grid=3, k=3, noise_scale=1.0)
f = lambda x: tf.math.exp(tf.math.sin(np.pi * x[:, 0:1]) + tf.math.square(x[:, 1:2]))
dataset = create_dataset(f, n_var=2, train_num=1000)


# y = model(dataset['train_input'])
history = model.fit(dataset, opt="LBFGS", steps=20, lamb=0.001, lr=0.0001,display_metrics=None)

grids = [5,10,20, 50]
for i in range(len(grids)):
    model=model.refine(grids[i])
    results = model.fit(dataset, opt="LBFGS",steps=20, stop_grid_update_step=20,lr=0.0001)
    history['train_loss'].extend(results['train_loss'])
    history['test_loss'].extend(results['test_loss'])
    history['reg'].extend(results['reg'])
model.plot()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(history['train_loss'], label="Train Loss", marker='o', linestyle='-')
plt.plot(history['test_loss'], label="Test Loss", marker='s', linestyle='--')
#plt.plot(results_history['reg'], label="Regularization", marker='^', linestyle=':')
plt.xlabel('Steps')
plt.ylabel('Values')
plt.yscale('log')
plt.title('Loss and Regularization Over Training Steps for KAN')
plt.legend()
plt.grid()
plt.show()
```

```python
from KERAS_KAN.kan.KANLayer import KANLayer
model = tf.keras.models.Sequential([
    KANLayer(in_dim=2, out_dim=5, ), 
    KANLayer(in_dim=5, out_dim=1, )   
])

# Build the model with an example input shape
model.build(input_shape=(None, 2))

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-2), 
    loss='mse', metrics=['mse'])
x_train = dataset['train_input']
y_train = dataset['train_label']
history_keras_kan = model.fit(x_train, y_train,  epochs=50, batch_size=64, validation_data=(dataset['test_input'],dataset['test_label']))
```

```python
plt.figure(figsize=(8, 5))
plt.plot(history_keras_kan.history['loss'], label='Train Loss', marker='o', linestyle='-')
plt.plot(history_keras_kan.history['val_loss'], label='Validation Loss', marker='s', linestyle='--')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.yscale('log')
plt.title('Training and Validation Loss Curves Keras Kan')
plt.legend()
plt.grid(True)
plt.show()
```

```python
model_mlp = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(2,)),  
    tf.keras.layers.Dense(64, activation='relu'),                    
    tf.keras.layers.Dense(1)                   
])


model_mlp.compile(
              optimizer=tf.keras.optimizers.Adam(learning_rate=1e-2),
              loss=tf.keras.losses.MeanSquaredError(),  
              metrics=['mse'])

history_mlp = model_mlp.fit(x_train, y_train, epochs=50, batch_size=64,validation_data=(dataset['test_input'],dataset['test_label']))
```

```python
plt.figure(figsize=(8, 5))
plt.plot(history_mlp.history['loss'], label='Train Loss', marker='o', linestyle='-')
plt.plot(history_mlp.history['val_loss'], label='Validation Loss', marker='s', linestyle='--')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.yscale('log')
plt.title('Training and Validation Loss Curves for MLP')
plt.legend()
plt.grid(True)
plt.show()
```

## Step 2:
$$f(\theta, \sigma) = \frac{\exp\left(-\frac{\theta^2}{2\sigma^2}\right)}{\sqrt{2\pi\sigma^2}}$$

```python
from tensorkan.utils import create_dataset
import numpy as np
from tensorkan.MultKAN import MultKAN as KANTF
import tensorflow as tf


f = lambda x: (1 / tf.sqrt(2 * np.pi * x[:, 1:2]**2)) * \
              tf.exp(-x[:, 0:1]**2 / (2 * x[:, 1:2]**2))


dataset = create_dataset(f, n_var=2, ranges=[[-3, 3], [0.1, 2.0]], train_num=1000, test_num=200)


model = KANTF(width=[2, 2, 1, 1], grid=3, k=3, noise_scale=1.0)


model.fit(dataset, opt="LBFGS", steps=20, lamb=0.001, lr=0.0001, display_metrics=None)

grids = [5,10,20, 50]
for i in range(len(grids)):
    model=model.refine(grids[i])
    model.fit(dataset, opt="LBFGS",steps=20, stop_grid_update_step=20,lamb=0.001,lr=0.0001)
model.plot()
```

```python
from KERAS_KAN.kan.KANLayer import KANLayer
model = tf.keras.models.Sequential([
    KANLayer(in_dim=2, out_dim=2, ),
    KANLayer(in_dim=2, out_dim=1, ),
    KANLayer(in_dim=1, out_dim=1, )  
])

# Build the model with an example input shape
model.build(input_shape=(None, 2))

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-2), 
    loss='mse', metrics=['mse'])
x_train = dataset['train_input']
y_train = dataset['train_label']
history = model.fit(x_train, y_train, epochs=100, batch_size=64,validation_data=(dataset['test_input'],dataset['test_label']))
```

```python
model_mlp = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(2,)),  
    tf.keras.layers.Dense(64, activation='relu'),                    
    tf.keras.layers.Dense(1)                   
])


model_mlp.compile(
              optimizer=tf.keras.optimizers.Adam(learning_rate=1e-2),
              loss=tf.keras.losses.MeanSquaredError(),  
              metrics=['mse'])

history = model_mlp.fit(x_train, y_train, epochs=100, batch_size=64,validation_data=(dataset['test_input'],dataset['test_label']))
```

## Step 3:
$$f(\theta, \theta_1, \sigma) = \frac{\exp\left(-\frac{(\theta - \theta_1)^2}{2\sigma^2}\right)}{\sqrt{2\pi\sigma^2}}$$

```python
f = lambda x: (1 / tf.sqrt(2 * np.pi * x[:, 2:3]**2)) * \
              tf.exp(-((x[:, 0:1] - x[:, 1:2])**2) / (2 * x[:, 2:3]**2))


dataset = create_dataset(f, n_var=3, ranges=[[-3, 3], [-3, 3], [0.1, 2.0]], train_num=1000, test_num=200)


model = KANTF(width=[3,2,2,1,1], grid=3, k=3, noise_scale=1.0)

model.fit(dataset, opt="LBFGS", steps=20, lamb=0.001, lr=0.0001, display_metrics=None)

grids = [5,10,20, 20]
for i in range(len(grids)):
    model=model.refine(grids[i])
    model.fit(dataset, opt="LBFGS",steps=20, stop_grid_update_step=20,lamb=0.001,lr=0.0001)
model.plot()
```

```python
from KERAS_KAN.kan.KANLayer import KANLayer
model = tf.keras.models.Sequential([
    KANLayer(in_dim=3, out_dim=4, ),
    KANLayer(in_dim=4, out_dim=1, ),
    #KANLayer(in_dim=2, out_dim=1, ) ,
    #KANLayer(in_dim=1, out_dim=1, )
])

# Build the model with an example input shape
model.build(input_shape=(None, 3))

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-2), 
    loss='mse', metrics=['mse'])
x_train = dataset['train_input']
y_train = dataset['train_label']
history = model.fit(x_train, y_train, epochs=100, batch_size=64)
```

```python
model_mlp = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(3,)),  
    tf.keras.layers.Dense(64, activation='relu'),                    
    tf.keras.layers.Dense(1)                   
])


model_mlp.compile(
              optimizer=tf.keras.optimizers.Adam(learning_rate=1e-2),
              loss=tf.keras.losses.MeanSquaredError(),  
              metrics=['mse'])

history = model_mlp.fit(x_train, y_train, epochs=100, batch_size=64)
```

## Step 4:
$$p = \sqrt{1 + a^2 - 2a \cos(\theta_1 - \theta_2)}$$

```python
f = lambda x: tf.sqrt(1 + x[:, 0:1]**2 - 2 * x[:, 0:1] * tf.cos(x[:, 1:2] - x[:, 2:3]))


dataset = create_dataset(f, n_var=3, ranges=[[0.1, 2.0], [-np.pi, np.pi], [-np.pi, np.pi]], 
                         train_num=1000, test_num=200)


model = KANTF(width=[3, 2, 2 , 1], grid=3, k=3, noise_scale=1.0)


model.fit(dataset, opt="LBFGS", steps=20, lamb=0.001, lr=0.0001, display_metrics=None)
grids = [5,10,20, 20]
for i in range(len(grids)):
    model=model.refine(grids[i])
    model.fit(dataset, opt="LBFGS",steps=20, stop_grid_update_step=20,lr=0.0001)
model.plot()
```

```python
from KERAS_KAN.kan.KANLayer import KANLayer
model = tf.keras.models.Sequential([
    KANLayer(in_dim=3, out_dim=2, ),
    KANLayer(in_dim=2, out_dim=2, ),
    KANLayer(in_dim=2, out_dim=1, ) ,
    #KANLayer(in_dim=1, out_dim=1, )
])

# Build the model with an example input shape
model.build(input_shape=(None, 3))

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-2), 
    loss=tf.keras.losses.MeanSquaredError(), metrics=['mse'])
x_train = dataset['train_input']
y_train = dataset['train_label']
history = model.fit(x_train, y_train, epochs=150, batch_size=64)
```

```python
model_mlp = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(3,)),  
    tf.keras.layers.Dense(64, activation='relu'),                    
    tf.keras.layers.Dense(1)                   
])


model_mlp.compile(
              optimizer=tf.keras.optimizers.Adam(learning_rate=1e-2),
              loss=tf.keras.losses.MeanSquaredError(),  
              metrics=['mse'])

history = model_mlp.fit(x_train, y_train, epochs=100, batch_size=64)
```
