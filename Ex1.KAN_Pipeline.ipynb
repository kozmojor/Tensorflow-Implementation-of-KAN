# Experiment 1: Welcome to Our KAN Reproduction Results！

From this on, we will show in this repo our experiments on the KAN we build using Tensorflow 2.14.0. This one will give you a first look at the basic Pipeline of KAN, and let's see what happens!

**Objective**: A demonstration on how KAN approximates a math formula!

## Step 1: Import Libraries and Set Up Environment and First Shot

We start by importing the required libraries and setting up the experiment environment. And then we'll get dataset and use what we get to train a basic KAN model. This single block is the core of KAN, and the marvellous picture below we get is nothing we have witnessed in classical MLP models.

```python
from tensorkan.utils import create_dataset
import numpy as np
from tensorkan.MultKAN import MultKAN as KANTF
import tensorflow as tf
from tensorkan.KANLayer import KANLayer  

model = KANTF(width=[2, 5, 1], grid=3, k=3, noise_scale=1.0)
f = lambda x: tf.math.exp(tf.math.sin(np.pi * x[:, 0:1]) + tf.math.square(x[:, 1:2]))
dataset = create_dataset(f, n_var=2, train_num=1000)

# y = model(dataset['train_input'])
model.fit(dataset, opt="LBFGS", steps=40, lamb=0.001, lr=0.0001);
# model.fit(dataset, opt="Adam", steps=10);
model.plot()
```

Wow, how decent the picture is! How KAN it manage to do so? Once you go KAN, you KAN never go back to sigmoid again... Well, that's a joke, we'll talk about KAN versus MLP in later experiments.

## Step 2: Prune is All You Need

The smart KAN has sorted out the 'neurons' we need at this step, we simply prune on KAN model and only keep the ones we need... and train again.

```python
model = model.prune()
model.plot()
```

```python
model.fit(dataset, opt="LBFGS", steps=50, lamb=0.001, lr=0.0001);
```

```python
model.plot()
```

## Step 3: Refine & Better

Well, KAN is not just drawing decent pictures and nothing further. We state that by the following step of Refine, KAN gains the ability to approximate any formulas in the field of Kolmogorov Arnold Representation Theorem.

```python
model = model.refine(10)
```

```python
model.fit(dataset, opt="LBFGS", steps=50);
```

So it's shown that both approximation loss and generalization loss improved. AND the picture shows more nonlinear funtions!

```python
model.plot()
```

## Step 4: Formula Pridiction

And KAN is something more, much more than what we showed above. With great ability of approximation, KAN can rebuild the formula it learns as below... And the accuracy will be even higher if loss keeps droping, and the formula rebuilt will gradually converge to the real one!

```python
from kan.utils import ex_round

ex_round(model.symbolic_formula()[0][0],4)
```
