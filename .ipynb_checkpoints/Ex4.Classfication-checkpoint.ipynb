# Example 4: Classification

Can KAN replace MLP? Before answering this question, the most important issue is: experiments. We need to see how KAN performes in the experiments that MLP are good at. And Classification is one of these

**Objective**: KAN for classification

## Step 1: Regression formulation

We first treat the problem as a regression problem (output dimension = 1, MSE loss). 

create the two moon dataset

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from tensorkan.utils import create_dataset
import numpy as np
from tensorkan.MultKAN import MultKAN as KANTF
import tensorflow as tf
from tensorkan.KANLayer import KANLayer

# Set the device to GPU if available, otherwise use CPU
device = '/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'
print(device)
# Initialize an empty dictionary for the dataset
dataset = {}

# Create training and testing datasets using the make_moons function
train_input, train_label = make_moons(n_samples=1000, shuffle=True, noise=0.1, random_state=None)
test_input, test_label = make_moons(n_samples=1000, shuffle=True, noise=0.1, random_state=None)

# Convert datasets to TensorFlow tensors and allocate them to the specified device
with tf.device(device):
    dataset['train_input'] = tf.convert_to_tensor(train_input, dtype=tf.float32)
    dataset['test_input'] = tf.convert_to_tensor(test_input, dtype=tf.float32)
    dataset['train_label'] = tf.convert_to_tensor(train_label[:, None], dtype=tf.float32)
    dataset['test_label'] = tf.convert_to_tensor(test_label[:, None], dtype=tf.float32)

# Extract training data for visualization
X = dataset['train_input']
y = dataset['train_label']

# Visualize the dataset using a scatter plot
plt.scatter(X[:, 0].numpy(), X[:, 1].numpy(), c=y[:, 0].numpy())
plt.show()
```

Train KAN

```python
from tensorkan.utils import create_dataset
import numpy as np
from tensorkan.MultKAN import MultKAN as KANTF
import tensorflow as tf

model = KANTF(width=[2,1], grid=3, k=3, noise_scale=1.0)

def train_acc():
        predictions = tf.round(model(dataset['train_input'])[:, 0])
        # Compare predictions with true labels and compute the mean accuracy
        accuracy = tf.reduce_mean(tf.cast(tf.equal(predictions, dataset['train_label'][:, 0]), dtype=tf.float32))       
        return accuracy

def test_acc():
    # Get predictions by rounding the model outputs
        predictions = tf.round(model(dataset['test_input'])[:, 0])
        # Compare predictions with true labels and compute the mean accuracy
        accuracy = tf.reduce_mean(tf.cast(tf.equal(predictions, dataset['test_label'][:, 0]), dtype=tf.float32))
        return accuracy

results = model.fit(dataset, opt="LBFGS", lamb=0.001, lr=0.0001, steps=20, metrics=(train_acc, test_acc));
results['train_acc'][-1], results['test_acc'][-1]
```

Apply automatic symbolic regression

```python
from tensorkan.utils import ex_round

lib = ['x','x^2','x^3','x^4','exp','log','sqrt','tanh','sin','tan','abs']
model.auto_symbolic(lib=lib)
formula = model.symbolic_formula()[0][0]
ex_round(formula, 4)
```

We now test the accuracy for this formula.

```python
def acc(formula, X, y):
    batch = X.shape[0]
    correct = 0
    for i in range(batch):
        pred = np.round(np.array(formula.subs('x_1', X[i,0]).subs('x_2', X[i,1])).astype(np.float64))
        correct += int(pred == y[i,0])
    return correct / batch

print('train acc of the formula:', acc(formula, dataset['train_input'], dataset['train_label']))
print('test acc of the formula:', acc(formula, dataset['test_input'], dataset['test_label']))
```

## Step 2: Classification formulation

We then treat the problem as a classification problem (output dimension = 2, CrossEntropy loss). 

Create the two moon datatset

```python
dataset = {}
train_input, train_label = make_moons(n_samples=1000, shuffle=True, noise=0.1, random_state=None)
test_input, test_label = make_moons(n_samples=1000, shuffle=True, noise=0.1, random_state=None)

dataset['train_input'] = tf.convert_to_tensor(train_input, dtype=tf.float32)
dataset['test_input'] = tf.convert_to_tensor(test_input, dtype=tf.float32)
dataset['train_label'] = tf.convert_to_tensor(train_label, dtype=tf.int32)
dataset['test_label'] = tf.convert_to_tensor(test_label, dtype=tf.int32)

X = dataset['train_input']
y = dataset['train_label']

plt.scatter(X[:, 0].numpy(), X[:, 1].numpy(), c=y.numpy(), cmap='viridis')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Training Data - Make Moons")
plt.show()
```

Train KAN

```python
model = KANTF(width=[2, 2], grid=3, k=3, seed=2024)

# Preprocess dataset labels to ensure they are of type int32
dataset['train_label'] = tf.cast(tf.squeeze(dataset['train_label']), dtype=tf.int32)
dataset['test_label'] = tf.cast(tf.squeeze(dataset['test_label']), dtype=tf.int32)

# Define a function to compute training accuracy
def train_acc():
    logits = model(dataset['train_input'])  # Get the model's predictions on training data
    logits = tf.cast(logits, dtype=tf.float32)  # Ensure logits are of type float32
    predictions = tf.cast(tf.argmax(logits, axis=1), dtype=tf.int32)   # Convert logits to predicted classes
    accuracy = tf.reduce_mean(tf.cast(predictions == dataset['train_label'], dtype=tf.float32))
    return accuracy

# Define a function to compute testing accuracy
def test_acc():
    logits = model(dataset['test_input'])  # Get the model's predictions on test data
    logits = tf.cast(logits, dtype=tf.float32)  # Ensure logits are of type float32
    predictions = tf.cast(tf.argmax(logits, axis=1), dtype=tf.int32)   # Convert logits to predicted classes
    accuracy = tf.reduce_mean(tf.cast(predictions == dataset['test_label'], dtype=tf.float32))
    return accuracy

# Apply training
results = model.fit(
    dataset,
    opt="LBFGS",
    steps=5,
    metrics=(train_acc, test_acc),
    loss_fn="SparseCategoricalCrossentropy"
)
model.plot()
```

Apply automatic symbolic regression

```python
lib = ['x','x^2','x^3','x^4','exp','log','sqrt','tanh','sin','abs']
model.auto_symbolic(lib=lib)
```

```python
from tensorkan.utils import ex_round
formula1, formula2 = model.symbolic_formula()[0]
ex_round(formula1, 4)
```

```python
ex_round(formula2, 4)
```

```python
model.plot()
```

We test the accuracy for this formula.

```python
def acc(formula1, formula2, X, y):
    batch = X.shape[0]
    correct = 0
    for i in range(batch):
        logit1 = np.array(formula1.subs('x_1', X[i,0]).subs('x_2', X[i,1])).astype(np.float64)
        logit2 = np.array(formula2.subs('x_1', X[i,0]).subs('x_2', X[i,1])).astype(np.float64)
        correct += int((logit2 > logit1) == y[i])
    return correct/batch

print('train acc of the formula:', acc(formula1, formula2, dataset['train_input'], dataset['train_label']))
print('test acc of the formula:', acc(formula1, formula2, dataset['test_input'], dataset['test_label']))
```

## Step 3: Analysis of Results

From this experiment, it is shown that KAN has the capability of handling tasks in different scenarios. Nevertheless, analysis points out some factors which should be considered in classification applications: 

### Network Depth
Shallow networks, like the two-layer KAN used here, can perform well on relatively simple tasks such as the two-moon dataset. However, for more complex or hierarchical data, deeper networks are needed to capture compositional structures effectively. Deeper networks can achieve better expressiveness, whereas wider shallow networks may reach their limits on such tasks.

### Grid Size and Complexity of Activation Functions
Increasing the grid size and depth of the KAN improves accuracy and allows the model to better approximate complex decision boundaries. However, this comes at the cost of higher computational complexity and longer training times, which need to be carefully managed depending on the application.

### Noise and Data Variability
The presence of noise in the data can affect the network’s performance. While KAN performs well on noisy datasets (as shown in the two-moon experiment), careful regularization or fine-tuning may be needed for more challenging data distributions.
